```python
from glob import glob
import os
from PIL import Image
import torch
import numpy as np
```

```python
dataset_dir = '../stimuli'
file_list = glob(os.path.join(dataset_dir, 'pathways*.jpg'))
len(file_list)
```

```python
img = Image.open(file_list[0])
img
```

```python
img.resize((512, 512))
```

```python
img_torch = torch.tensor(np.array(img)).permute(2,0,1)
```

```python
img_torch.shape
```

```python
from torchvision.transforms import Resize
```

```python
resize_op = Resize(size=(512, 512))
resized_im = resize_op(img_torch)
resized_im.shape
```
